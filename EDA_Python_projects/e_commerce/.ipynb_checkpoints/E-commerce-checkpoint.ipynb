# Count the total number of unique customers in January 2011 and how many of them came back every month over the entire year in 2011 

```python
df.head().T
```

```python
df[(df["Order_Date"].dt.year == 2011) & (df["Order_Date"].dt.month == 1)].head()
```

```python
df[(df["Order_Date"].dt.year == 2011) & (df["Order_Date"].dt.month == 1)][["Cust_id"]].nunique()
```

```python
import calendar
```

```python
my_dict = {}
for i in list(range(1, 13)):
    customers_2011jan = df[(df["Order_Date"].dt.year == 2011) & (df["Order_Date"].dt.month == 1)]["Cust_id"]
    if i == 1:
        print("count of customers in Jan 2011 is", customers_2011jan.nunique())        
        print("--------------")
    else: 
        a = df[(df["Order_Date"].dt.year == 2011) & (df["Order_Date"].dt.month == i) & (df.Cust_id.isin(customers_2011jan))]["Cust_id"].nunique()
        print(round(a / customers_2011jan.nunique() * 100, 1),  "percent of them came back in", calendar.month_name[i])
        my_dict[calendar.month_name[i]] = round(a / customers_2011jan.nunique() * 100, 1)
print("--------------")
data = {"months_of_2011": my_dict.keys(), "came_back_ratio_after_Jan": my_dict.values()}
came_back_df = pd.DataFrame(data)
came_back_df
    
```

# for  each  user,  find  the  time  elapsed  between  the  first purchasing and the third purchasing, in ascending order by Customer ID

```python
df.head().T
```

```python
df.columns
```

```python
mydict = {}
name_dict = {}
for i in df.Cust_id.unique():
    a = df[df["Cust_id"] == i][["Order_Date"]].sort_values(by="Order_Date").drop_duplicates()
    if len(a) >= 3:
        b = (a.iloc[2] - a.iloc[0]).dt.days[0]
        mydict[i] = b
        name_dict[i] = df[df["Cust_id"] == i][["Customer_Name"]].values[0][0]
        
data = {"Cust_id": mydict.keys(), "Customer_Name" : name_dict.values(), "days_between1st_3th_purchase": mydict.values()}
cust_more3purchases = pd.DataFrame(data)           
cust_more3purchases.sort_values(by="days_between1st_3th_purchase")
    
        
```

# Find the customers who purchased both product 11 and product 14, as well as the ratio of these products to the total number of products purchased by the customer.

```python
df.head().T
```

```python
df11 = df[df["Prod_id"] == 11]
df14 = df[df["Prod_id"] == 14]
```

```python
df11and14 = df11[["Cust_id", "Prod_id", "Order_Quantity"]].merge(df14[["Cust_id", "Prod_id", "Order_Quantity"]], on="Cust_id", how = "inner")
```

```python
df11and14["total_11_14"] = df11and14["Order_Quantity_x"] + df11and14["Order_Quantity_y"]
```

```python
df11and14
```

```python
df11and14 = df11and14.groupby("Cust_id")[["total_11_14"]].sum()
df11and14.reset_index(inplace=True)
```

```python
df11and14
```

```python
list_cust_id_for_11_14 = list(df11and14["Cust_id"])
list_cust_id_for_11_14
```

```python
df.loc[df["Cust_id"].isin(list_cust_id_for_11_14)]
```

```python
df_order_count = df.loc[df["Cust_id"].isin(list_cust_id_for_11_14)].groupby("Cust_id")["Order_Quantity"].sum().reset_index()
df_order_count
```

```python
df11and14_and_total = df11and14.merge(df_order_count)
df11and14_and_total
```

```python
df11and14_and_total["ratio_11_14to_total"] = round((df11and14_and_total["total_11_14"] / df11and14_and_total["Order_Quantity"] * 100), 1)
df11and14_and_total.rename(columns = {"Order_Quantity": "total_quantitiy"}, inplace=True)
df11and14_and_total
```
